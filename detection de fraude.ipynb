## Performing a K-means Clustering

```python
df_acp.head()
```

```python
from sklearn.cluster import KMeans
```

```python
# Create a k-means clustering model
kmeans = KMeans(n_clusters=2, n_init=10)

# fit the data to the model
kmeans.fit(x_scaled)

# Determine which clusters each data point belongs to:
clusters = kmeans.predict(x_scaled)
```

```python
# clusters associés pour chaque individu - méthode 1
clusters
```

```python
# clusters associés pour chaque individu - methode 2
kmeans.labels_
```

```python
x_scaled_clustered = pd.DataFrame(x_scaled, columns=df_acp.columns, index=df_acp.index)
x_scaled_clustered.head(3)
```

```python
# associons chaque individu du df centré-réduit au cluster qui lui est associé
x_scaled_clustered['cluster'] = clusters
x_scaled_clustered.head()
```

```python
df.head()
```

### Regression logistique

```python
df.corr()['is_genuine']
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
reg_log = smf.glm('is_genuine ~ margin_low + length',
                   data=df_converted, family=sm.families.Binomial()).fit()
print(reg_log.summary())
```

```python
import statsmodels.formula.api as smf
model = smf.logit('is_genuine ~ margin_low + length', data=df_converted)
results = model.fit()
print(results.summary())
```

```python
predicted_class_prob = results.predict(df_converted[['length', 'margin_low']])
predicted_class_prob.head()
```

```python
threshold = 0.5
# transforms the list of booleans into an int with 1 = True, 0 = False
predicted_class = (predicted_class_prob > threshold).astype(int)
predicted_class.head()
```

```python
results.pred_table()
# 69 faux billets et 99 vrais billets ont été correctement catologué comme faux et vrais billets
# 1 faux billet a été catalogué comme vrai billet et un vrai billet a été classé comme faux billet
```

```python
sns.distplot(predicted_class_prob)
plt.title('distribution true and false value')
plt.xlabel('probability')
plt.show()
```
